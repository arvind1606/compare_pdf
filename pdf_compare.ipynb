# Compare two PDF's using pdfplumber

```python
import pdfplumber
import difflib

def compare_pdfs(pdf_file1, pdf_file2):
    # Open the first PDF file
    with pdfplumber.open(pdf_file1) as pdf1:
        # Open the second PDF file
        with pdfplumber.open(pdf_file2) as pdf2:
            # Extract text content from both PDFs
            text1 = " ".join([page.extract_text() for page in pdf1.pages])
            text2 = " ".join([page.extract_text() for page in pdf2.pages])

            # Split text into lines for comparison
            lines1 = text1.splitlines()
            lines2 = text2.splitlines()

            # Perform a line-by-line comparison
            differ = difflib.Differ()
            diff = list(differ.compare(lines1, lines2))

            return diff

if __name__ == "__main__":
    pdf_file1 = "ESHRE 2023 Email Blast v2.pdf"  # Replace with your first PDF file
    pdf_file2 = "ESHRE 2023 Email Blast v3.pdf"  # Replace with your second PDF file

    differences = compare_pdfs(pdf_file1, pdf_file2)

    # Print the differences
    for line in differences:
        print(line)
```
